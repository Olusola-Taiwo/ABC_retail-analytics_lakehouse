## The Transformation Logic

```python
from datetime import datetime

# ---------------------------------------------------------
# 1. Build today's suffix once
# ---------------------------------------------------------
today = datetime.today().strftime("%Y_%m_%d")

# ---------------------------------------------------------
# 2. Build the two Silver tables used in THIS query
# ---------------------------------------------------------
crm_products = f"`abc_business-core`.silver.crm_products_{today}"
erp_product_category = f"`abc_business-core`.silver.erp_pricing_{today}"

# ---------------------------------------------------------
# 3. Adapted SQL query with dynamic table names
# ---------------------------------------------------------
query = f"""
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.start_date, pn.product_number) AS product_key,
    pn.product_id,
    pn.product_number,
    pn.product_name,
    pn.category_id,
    pc.category,
    pc.subcategory,
    pc.maintenance_flag,
    pn.product_line,
    pn.start_date
FROM {crm_products} pn
LEFT JOIN {erp_product_category} pc
    ON pn.category_id = pc.category_id
"""

# ---------------------------------------------------------
# 4. Execute
# ---------------------------------------------------------
df = spark.sql(query)
```
