## The Transformation Logic

```python
from datetime import datetime

# ---------------------------------------------------------
# 1. Build today's suffix once
# ---------------------------------------------------------
today = datetime.today().strftime("%Y_%m_%d")

# ---------------------------------------------------------
# 2. Build the Silver + Gold tables used in THIS query
# ---------------------------------------------------------
crm_sales = f"`abc_business-core`.silver.crm_sales_{today}"
gold_dim_products = f"`abc_business-core`.gold.dim_products_{today}"
gold_dim_customers = f"`abc_business-core`.gold.dim_customers_{today}"

# ---------------------------------------------------------
# 3. Adapted SQL query with dynamic table names
# ---------------------------------------------------------
query = f"""
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM {crm_sales} sd
LEFT JOIN {gold_dim_products} pr
    ON sd.product_number = pr.product_number
LEFT JOIN {gold_dim_customers} cu
    ON sd.customer_id = cu.customer_id
"""

# ---------------------------------------------------------
# 4. Execute
# ---------------------------------------------------------
df = spark.sql(query)
display(df)
```
